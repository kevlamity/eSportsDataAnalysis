## Aims and background
### Introduction to the game [0]
Counter-Strike:Global Offensive is a team based 5 vs 5 competitve FPS (first-person shooter) game. 

Each round of a match involves 5 Counter-terrorists (CTs) taking on 5 terrorists (Ts) where the objective of the Ts are to plant and detonate a bomb at either of the 2 bombsites on the map. They can also win rounds by eliminating all 5 CTs.

The objectives of the CTs are to stop the Ts from planting and detonating the bomb at either bombsites, either by eliminating all 5 of them, defusing the planted bomb, or running down the round timer of 1min 55seconds.

Each match spans a total of 30 rounds, with teams playing 15 rounds of a side first before switching to the other. The match is won when either team reaches 16 rounds first. If both team gets 15 rounds, overtime is played which is best of 6 rounds(first to 4 wins), which can continue indefinitely if teams continuously tie (15-15 is 1st overtime, 18-18 is 2nd overtime etc.).

```python
#request page function, with player/webpage parameter as the argument
def request_page(player):
    S = requests.Session()
    PARAMS = {
    "action": "parse",
    "page": player,
    "format": "json"
    }
    URL = "https://liquipedia.net/counterstrike/api.php"
    #Including headers that liquipedia's API terms require such as including gzip encoding
    headers = {"User-Agent":"Chrome Version 102.0.5005.115 / Windows NT 10.0; Win64; x64; rv:66.0 Python3",
           "Upgrade-Insecure-Requests":"1", 
           "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", 
           "Accept-Encoding":"gzip, deflate, br", 
           "Accept-Language": "en-GB,en-US;q=0.9,en;q=0.8",
           "DNT":"1"}
    player_page = requests.get(url=URL, params=PARAMS, headers=headers)
    player_page.raise_for_status()
    DATA = player_page.json()
    return(DATA)
```

### Dealing with Redirect pages
Some player names on HLTV.org do not match exactly on liquipedia.net, thus there is a redirect page on liquipedia.net for different name variations. This is solved by checking if the first page request is a redirect page, then getting the correct naming convention of the player(the redirect page shows the player's correct naming convention) e.g "device redirects to dev1ce"

### Dealing with Disambiguation pages
There are also instances of players having the same game nickname, which are differentiated by their nationality on liquipedia for the url query. For example, the player name "Lucky" has multiple players using it, with the url https://liquipedia.net/counterstrike/Lucky bringing us to a disambiguation page showing the multiple players with the name. These players' wiki page are thus differentiated by their nationality; the french player Lucky wiki page would be https://liquipedia.net/counterstrike/Lucky_(French_player)

```python
# Getting individual liquipedia(wiki of players) page of players
for i in range(len(playersstats_df)):
    try:
        playername = playersstats_df.iloc[i][0]
        
        redirect = BeautifulSoup(request_page(playername)["parse"]["text"]["*"], "html.parser")
        #Checks if the player page requested is a Disambiguation page
        if(redirect.find("a").get('title') == "Category:Disambiguation pages"):
            player_nationality = playersstats_df.iloc[i][7]
            print(player_nationality)
            playername = playername + "_(" + player_nationality + "_player)"
            playersstats_df.iloc[i][0] = playername
            player_details(request_page(playername))
        #If player page is not a Disambiguation page and not a redirect page, parse the player page for their details
        elif(redirect.find("p").text != "Redirect to:"):
            try:
                if(search("For the", redirect.find("span",class_="navigation-not-searchable").findNext("span").findNext("i").text) or search("For the", redirect.find("span",class_="navigation-not-searchable").findNext("i").text) ):
                    player_nationality = playersstats_df.iloc[i][7]
                    for row in redirect.find_all("span"):
                        try:
                            if(search(player_nationality,row.findNext("i").text)):
                                nationality_str = row.find("a").get("title").replace(' ', '_')
                        except Exception:
                            pass
                    print(nationality_str)
                    playername = nationality_str
                    playersstats_df.iloc[i][0] = str(playername)
                    player_details(request_page(playername))
                else:
                    player_details(request_page(playername))
            except Exception:
                player_details(request_page(playername))
        #If the first page requested is a redirect page, proceed to the redirect page and get the playername's correct naming convention
        elif(redirect.find("p").text == "Redirect to:"):
            playername = redirect.find("a").text.replace(' ', '_')
            #changing player's name to liquipedia's naming convention so that earnings page can be easily accessed later
            playersstats_df.iloc[i][0] = playername
            player_details(request_page(playername))
        
    # catch any errors
    except HTTPError:
        print (player_page.status_code, 'Request failed')
    else:
        print ('Request successful')
        #50 Second sleep timer for each player request, to prevent getting blocked/banned from using Liquipedia's API
        time.sleep(50)
```

### Success Rating
To measure a player's success for professional CS:GO, two factors come into mind. Their rating in-game and their earnings. Looking at either individually could give us a sense of which are the most successful players, but might not accurately show a player's impact and success they have.

For example, players who perform well in-game with a high rating does not directly translate to their success at the top level of play. We have to look at how a player's rating affect their earnings for these 3 years of play.

A player with a high rating but low earnings could mean that they are "farming" low tier competition or only play well against lesser teams, such as playing well against lower rated teams/smaller tournaments thus having lesser earnings but a higher rating. This applies vice versa.

Adding to the fact that CS:GO is heavily focused on team-play, a player's rating does not tell the whole story of how successful a player is.

Thus, we need to create a better variable to determine a player's success, which includes both their in-game rating and earnings.

- Older players played the least amount of game rounds in the 3 years analysed(2019-2021), which might be due to the fact they have more life commitments than younger players
- The region with the youngest players and play the most rounds is CIS

### Project Conclusion
As CS:GO is a team game with many different roles and aspects to how games are won, it is difficult to conclude the ideal variables that dictate a players success. 

On one hand younger players might have higher ratings due to better mechanical game skill that might lead them to win games by kills, older players with lesser ratings might have more experience playing for a longer period and might lead them to win games objectively.

Since rounds can be won not just by killing opponents(which increases rating), it is entirely possible for winning teams to have a lower mean rating versus the losing team as they are able to win many rounds through playing the objective and not just hunting for kills or "stat padding" as they call it in FPS games.
